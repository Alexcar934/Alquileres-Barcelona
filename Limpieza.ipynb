```python
import pandas as pd
import re
```

```python
df = pd.read_csv('AlquileresBCN.csv', index_col=0)
df = df[df['Precio']!='No veo el precio'].drop_duplicates().reset_index(drop=True)
```

### Precios a enteros

```python
df['Precio'] = [int(i.replace('.','').replace('€','')) for i in df['Precio']]
df['Precio Antes'] = [int(i.replace('.','').replace('€','')) for i in df['Precio Antes']]
```

### Habitaciones

```python
habitaciones = []
for i in df['Habitaciones']:
    if i == 'No veo las habitaciones':
        habitaciones.append(0)
    else:
        habitaciones.append(int(i))
df['Habitaciones'] = habitaciones
df
```

### Baños

```python
df.Baños.unique()
```

```python
baños = []
for i in df['Baños']:
    if i == 'No encuentro los baños':
        baños.append(0)
    else:
        baños.append(int(i))
df['Baños'] = baños
df
```

### Metros Cuadrados

```python
df['Metros cuadrados'] = [int(i.split()[0]) for i in df['Metros cuadrados']]
```

### Columnas binarias

```python
df['Aire acondicionado'].unique()
```

```python
columnas_binarias = ['Aire acondicionado', 'Amueblado',
       'Armarios empotrados', 'Ascensor', 'Balcón', 'Calefacción', 'Terraza']

for i in columnas_binarias:
   lista = []
   for j in df[i]:
       if j == 'Sí':
              lista.append(1)
       else:
              lista.append(0)
   df[i] = lista    
            
```

### Características

Sacar la planta en la que se encuentra:
* 0 si no se especifica planta
* otros números son el número de la planta en la que se encuentra

```python
plantas = []
for i in df['Características']:
    if 'Planta' in i:
        pattern = re.compile(r'Planta \d+')
        matches = pattern.findall(i)

        for match in matches:
            plantas.append(match)
    else:
        plantas.append('0')

df['Planta'] = plantas
```

```python
df['Planta'] = [int(i.replace('Planta ','')) for i in df['Planta']]
```

```python
df['Planta'].unique()
```

```python
df['Planta'].value_counts()
```

Saber si es interior o exterior

* 0 si no lo especifica o es interior (hay pocos interiores)
* 1 si es exterior

```python
int_ext = []
for i in df['Características']:
    if ': exterior' in i:
        int_ext.append(1)
    else:
        int_ext.append(0)

df['Interior / Exterior'] = int_ext
```

```python
df['Interior / Exterior'].value_counts()
```

El Estado del piso:
* 0 si está por reformar o no especifica
* 1 si está en buen estado
* 2 si está nuevo

```python
estado = []
for i in df['Características']:
    if 'Estado' in i:
        pattern = re.compile(r"Estado:.[^,']+")
        matches = pattern.findall(i)

        for match in matches:
            estado.append(str(match).split(',')[0])
    else: 
        estado.append('No especifica')

df['Estado del piso'] = estado
```

```python
df['Estado del piso'].value_counts()
```

Calefacción:
* Le asignamos un valor del 1 al 5 en función del tipo de calefacción (0 si no especifica)

{'No especifica',
 'calefacción: central',
 'calefacción: electricidad',
 'calefacción: gas natural',
 'calefacción: gas propano',
 'calefacción: independiente'}

```python
calefacción = []
for i in df['Características']:
    if ': central' in i:
        calefacción.append(1)
    elif ': electricidad' in i:
        calefacción.append(2)
    elif ': gas natural' in i:
        calefacción.append(3)
    elif ': gas propano' in i:
        calefacción.append(4)
    elif ': independiente' in i:
        calefacción.append(5)
    else:
        calefacción.append(0)
```

```python
df['Sistema de calefacción'] = calefacción
```

```python
df['Sistema de calefacción'].value_counts()
```
