```python
import pandas as pd
```

```python
df = pd.read_csv('AlquileresBCN.csv', index_col=0)
df = df[df['Precio']!='No veo el precio'].drop_duplicates().reset_index(drop=True)
```

### Precios a enteros

```python
df['Precio'] = [int(i.replace('.','').replace('€','')) for i in df['Precio']]
df['Precio Antes'] = [int(i.replace('.','').replace('€','')) for i in df['Precio Antes']]
```

### Habitaciones

```python
habitaciones = []
for i in df['Habitaciones']:
    if i == 'No veo las habitaciones':
        habitaciones.append(0)
    else:
        habitaciones.append(int(i))
df['Habitaciones'] = habitaciones
df
```

### Baños

```python
df.Baños.unique()
```

```python
baños = []
for i in df['Baños']:
    if i == 'No encuentro los baños':
        baños.append(0)
    else:
        baños.append(int(i))
df['Baños'] = baños
df
```

### Metros Cuadrados

```python
df['Metros cuadrados'] = [int(i.split()[0]) for i in df['Metros cuadrados']]
```
